```python
#function to extract user timeline
def extract_timeline_as_df(timeline_list):
    columns = set()
    allowed_types = [str, int]
    tweets_data = []
    

    for status in timeline_list:
        
        status_dict = dict(vars(status))
        keys = vars(status).keys()
        single_tweet_data = {"user": status.user.screen_name, "author": status.author.screen_name}

        for k in keys:
            try:
                v_type = type(status_dict[k])
            except:
                v_type = None
            if v_type != None:
                if v_type in allowed_types:
                    single_tweet_data[k] = status_dict[k]
                    columns.add(k)
                    
        tweets_data.append(single_tweet_data)
    #print(columns)

    header_cols = list(columns)
    header_cols.append("user")
    header_cols.append("author")
    df = pd.DataFrame(tweets_data, columns = header_cols)
    return df 
```

```python
#Function to extract user data
def user_data(mylist):
    new_df = []
    for inv in mylist:
        try:
            user = api.get_user(inv)
            user_timeline = user.timeline(count=50, include_rts = False) # extract recent 50 tweets excluding retweets
            df = extract_timeline_as_df(user_timeline)
            new_df.append(df)
        except tweepy.TweepError:
            pass
    return new_df
```

```python
#function to create dataframe with user that has at least 10 tweets in english
def tweets_english(mylist, df):
    new_list = []
    #user with atleast 10 out of 50 tweets in english
    for inv in mylist:
        if (df[df.user==inv].lang == 'en').sum()>10:
            new_list.append(inv)
        else:
            continue
    #subset dataset using the above list
    new_df = df[df['user'].isin(new_list)]
    return new_df
```

```python
import time
#function to extract user data
def user_stat(mylist):
    user_list = []
    followers_count = []
    following_count = []
    status_count = []
    for ind in mylist:
        try:
            user = api.get_user(ind, wait_on_rate_limit=True)
            followers = user.followers_count
            following = user.friends_count
            statuses = user.statuses_count
            followers_count.append(followers)
            following_count.append(following)
            status_count.append(statuses)
            user_list.append(ind)
        except tweepy.TweepError:
            time.sleep(20)
    user_df = pd.DataFrame({'user':user_list,
                          'followers_count':followers_count,
                         'following_count':following_count,
                         'status_count':status_count})
    return user_df
```

## Target variables

Our target variable is education. Let's see the frequency distribution of target varaible.

```python
#bar plot of the target variable
sns.countplot(x='education', data = tweet_data)
plt.xlabel('Education level', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.show()
```

```python
#let's groupby 'favorite_count', 'retweet_count','Followers', 'Friends', 'Tweets' based on user
df_num = tweet_data.groupby('user')[['favorite_count', 'retweet_count','Followers', 'Friends', 'Tweets', 'education']].first()
df_num.head()
```

```python
#Check normality of tweet features
from scipy import stats
columns = ['favorite_count', 'retweet_count','Followers', 'Friends', 'Tweets']
for i in columns:
    print([i])
    a, b = stats.kstest(df_num[[i]], 'norm')
    print("statistics", a, "p-value", b)
    if b < 0.05:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject null hypothesis")
```

Kolmogorov-Smirnov test showed that p-value is less than 0.05 providing evidence to reject the null hypothesis and conclude the distributions are not normal. Next, we will perform log transformation to normalize the data.

```python
#log transform the data and check normality
columns = ['favorite_count', 'retweet_count','Followers', 'Friends', 'Tweets']
for i in columns:
    print([i])
    a, b = stats.kstest(np.log(df_num[[i]]+1), 'norm')
    print("statistics", a, "p-value", b)
    if b < 0.05:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject null hypothesis")
```

Even after log transformation, the data is still not normal. Next, we will perform boxcox transformation.

```python
#transform data with boxcox and check normality
from scipy.stats import boxcox
bc, lmda = boxcox(df_num['Tweets'])

for i in columns:
    print([i])
    bc_tra, lmda = boxcox(df_num['Tweets'])
    a, b = stats.kstest(bc_tra, 'norm')
    print("statistics", a, "p-value", b)
    if b < 0.05:
        print("Reject the null hypothesis")
    else:
        print("Fail to reject null hypothesis")
```

Data were transformed using log transformation and boxcox transformation; however, the data are still not normal. Therefore, the data were ranked and one-way ANOVA was carried out.

```python
#barplot of favorite count based on education
sns.set_style("whitegrid")
sns.catplot(x="education",y="favorite_count",data=df_num,kind="bar", capsize=.2, order=["phd","master", "bachelor", "hsg"]).set(title='Average favorite count')
plt.title('Average favorite count', fontsize=14)
plt.xlabel('Education level', fontsize=14)
plt.ylabel('Favorite count', fontsize=14)
plt.show()
```

```python
# rank the favorite_count
df_num['favorite_count_rank'] = df_num['favorite_count'].rank(method='max')
```

```python
import scipy.stats as stats
phd=df_num['education']=='phd'
master=df_num['education']=='master'
bachelor=df_num['education']=='bachelor'
hsg=df_num['education']=='hsg'

# compute and returns ANOVA F and p value
fvalue, pvalue = stats.f_oneway(df_num['favorite_count_rank'][phd],
                                df_num['favorite_count_rank'][master],
                                df_num['favorite_count_rank'][bachelor],
                                df_num['favorite_count_rank'][hsg])
print('F-Statistic=%.3f, p=%.3f' % (fvalue, pvalue))
```

The p value obtained from ANOVA analysis is significant (p < 0.05); therefore, it is concluded that there are significant differences among different education level for favorite_count.

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
# perform Tukey's test
tukey = pairwise_tukeyhsd(endog=df_num['favorite_count_rank'],
                          groups=df_num['education'],
                          alpha=0.05)

#display results
print(tukey)
```

Above results from Tukey’s HSD suggests that except phd-hsg and hsg-master, all other pairwise comparisons for education level rejects null hypothesis (p < 0.05) and indicates statistical significant differences.

## Text normalization

In the next step, I will convert the text column of the dataframe to list and then replace url with domain name, remove the tagged user name, replace emojis with text, convert text to lowercase, and remove numbers, punctutations, whitespaces, accented characters, and stop words. Then, I will perform lemmatization on the text.

```python
#text normalization
nlp = spacy.load("en_core_web_lg")
tokenizer = ToktokTokenizer()
stopword_list = nltk.corpus.stopwords.words('english')

class Text_Normalization:
    def __init__(self, word):
        self.word = word
    
    #replace weblink with _domainname_
    def domain_change(self):
        newstr = []
        for word in self.word.split(' '):
            extracted = tldextract.extract(word)
            if extracted.domain != '' and extracted.suffix != '':
                i = '_' + extracted.domain + '.' + extracted.suffix + '_'
            else:
                i = word + ' '
            newstr.append(i)
        new_st = ' '.join(newstr)
        self.new_st= new_st
        
    #function to remove tagging
    def no_tagging(self):
        no_tag = re.sub("@[\w]*", '', self.new_st)
        self.no_tag = no_tag
            
    #expand contractions
    def word_expand(self):
        expanded_words = []    
        for word in self.no_tag.split():
          # using contractions.fix to expand the shotened words
          expanded_words.append(contractions.fix(word))   
        expanded_text = ' '.join(expanded_words)
        self.expanded_text = expanded_text
        
    #lowercase
    def lower_case(self):
        low_case = self.expanded_text.lower()
        self.low_case = low_case
         
    #remove numbers
    def no_numbers(self):
        no_num = re.sub(r'\d+', '', self.low_case)
        self.no_num = no_num
        
    #remove punctuations
    def no_punct(self):
        punctuations = '!"$%&\'()*+,-./:;<=>?@[\\]^`{|}~'
        no_punctuation = self.no_num.translate(str.maketrans('', '', punctuations))
        self.no_punctuation = no_punctuation
        
    #replace emojis with text
    def emoji_to_word(self):
        noemoji = emoji.demojize(self.no_punctuation)
        noemoji = noemoji.replace(":","_")
        noemoji = ' '.join(noemoji.split())
        self.noemoji = noemoji
    
       
    #remove stop words
    def remove_stopwords(self, is_lower_case=False):
        tokens = tokenizer.tokenize(self.noemoji)
        tokens = [token.strip() for token in tokens]
        if is_lower_case:
            filtered_tokens = [token for token in tokens if token not in stopword_list]
        else:
            filtered_tokens = [token for token in tokens if token.lower() not in stopword_list]
        filtered_text = ' '.join(filtered_tokens)    
        self.filtered_text = filtered_text
        
    #replace everything not a letter or apostrophe with a space
    def only_alphabets(self):
        alphabets = re.sub(r'[^a-zA-Z ]+_#', '', self.filtered_text)
        self.alphabets = alphabets
    
    # Remove single letter words
    def no_single_char(self):
        multiletters = ' '.join( [w for w in self.alphabets.split() if len(w)>1])
        self.multiletters = multiletters
    
    #remove whitespaces
    def remove_whitespace(self):
        no_whitespace = self.multiletters.strip()
        self.no_whitespace = no_whitespace
           
    #remove accented characters
    def remove_accented_chars(self):
        no_acc_char = unicodedata.normalize('NFKD', self.no_whitespace).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        self.no_acc_char = no_acc_char
        
    
    def lemmatize_text(self):
        final_list = []
        for word in self.no_acc_char.split():
            if word.startswith("_") or word.startswith("#"):
                w = word
                final_list.append([w])
            else:
                text = nlp(word)
                w = [word.lemma_ if word.lemma_ != '-PRON-' else word.text for word in text]
                final_list.append(w)
        lemmatize_text = ' '.join([item for sublist in final_list for item in sublist])
        return lemmatize_text
```

```python
#convert column 'text' of the dataframe to list
my_list = tweet_data['text'].tolist()
```

```python
#normalize the tweets
normalized_tweets = []

for i in my_list:
    tn = Text_Normalization(i)
    urltodomain = tn.domain_change()
    no_tag = tn.no_tagging()
    word_ex = tn.word_expand()
    low_case = tn.lower_case()
    no_num = tn.no_numbers()
    no_punc = tn.no_punct()
    no_emoji = tn.emoji_to_word()
    #no_hashtag = tn.replace_hashtag()
    no_stopword = tn.remove_stopwords()
    only_alphabets = tn.only_alphabets()
    multichar = tn.no_single_char()
    no_whitespace = tn.remove_whitespace()
    no_acc_char = tn.remove_accented_chars()
    lemm = tn.lemmatize_text()
    normalized_tweets.append(lemm)
```

## Word embedding

If the words such as PhD or Master degree or bachelor or high school is in a tweet it is easy to predict the education level of such tweets. Therefore, it is better to delete such tweets. Here, I am using word2vec algorithm and spaCy's prebuilt word embeddings to find word similar to five words (degree, phd, master, bachelor, highschool). Then, I am deleting any tweets that contain these words or words that share more than 0.5 similarity with these words.

```python
#delete words related to class label ('degree', 'phd', 'master', 'bachelor', 'highschool')
wl = ['degree', 'phd', 'master', 'bachelor', 'highschool']
wd = '|'.join(r"\b{}\b".format(x) for x in wl)
tweet_data['tweet_normalized'] = tweet_data['tweet_normalized'].str.replace(wd, '')
```

```python
# let create a list of word that are similar to 'education degree phd master bachelor highschool'
nlp = spacy.load("en_core_web_lg")

# Create a word embedding using spaCy for the words related to education
embed_word = nlp('degree phd master bachelor highschool')
```

```python
# Iterate over  each tweet in the dataframe "tweet_normalized" to remove the above embed_word
for i, tweet in enumerate(tweet_data['tweet_normalized']):
    # Create a word embedding for the current tweet
    word_sim = nlp(tweet)
    
    # If the current tweet is similar to education by more than 0.5 then drop the tweet from the dataframe
    if word_sim.similarity(embed_word) > 0.5:
        tweet_data_embed = tweet_data.drop(i , inplace=True)
```
